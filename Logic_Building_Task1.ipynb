```python
#Question 2 - Pass / Fail Analyzer
```

```python
# List of student marks
marks = [45, 78, 90, 33, 60]

# Initialize counters to store the count of passed and failed students
pass_count = 0
fail_count = 0

# Analyze each student's marks
print("STUDENT RESULT ANALYSIS")
print("=" * 40)

for i, mark in enumerate(marks, 1):
    if mark >= 50:
        pass_count += 1
        status = "PASS"
    else:
        fail_count += 1
        status = "FAIL"

    # Print individual student result
    print(f"Student {i}: {mark} marks -> {status}")

# Display final summary
print(f"\nTotal Students: {len(marks)}")
print(f"Total Passed Students: {pass_count}")
print(f"Total Failed Students: {fail_count}")
```

```python
# Question 3: Simple Data Cleaner
```

```python
# Given list of names
names = [" Alice ", "bob", " CHARLIE "]

# Create an empty list to store cleaned names
updated_names = []

# Processing each name to remove spaces and converting to lower case
for name in names:
    updated_name = name.strip().lower()
    updated_names.append(updated_name)

# Print cleaned list
print("Original Names:", names)
print("Updated Names:", updated_names)
```

```python
# Question 4: Message Length Analyzer
```

```python
# List of messages
messages = ["Hi", "Welcome to the platform", "OK"]

print("MESSAGE ANALYSIS REPORT")
print("=" * 40)

#Identifying message size of each text
for index, text in enumerate(messages, start=1):
    message_length = len(text)
    
    print(f"Message {index}: \"{text}\"")
    print(f"Length: {message_length} characters")

    # Validating the message length
    if message_length > 10:
        print("Status: Long Message")
    else:
        print("Status: Short Message")
    
    print("-" * 40)
```

```python
# Question 5: Error Message Detector

# System log records
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Initialize counters to store "ERROR" counts
error_count = 0

# Validating the logs
for log_entry in logs:
    if log_entry == "ERROR":
        error_count += 1 

# Display final summary
print("LOG SUMMARY REPORT")
print("=" * 30)
print("Total log records:", len(logs))
print("Total ERROR occurrences:", error_count)
```
