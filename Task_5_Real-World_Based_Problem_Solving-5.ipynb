```python
# function to calculate likes 
def engagement_score(likes):
  total_likes = 0
  # Calculating total
  for i in likes:
    total_likes += i

  print("Total Likes:", total_likes)
  # Checking for the comdition 
  if total_likes >= 1000:
    print("Post Status: Viral Post")
  else:
    print("Post Status: Normal Engagement")
  return
    
likes = [50,100,200,300,400]
#function call
engagement_score(likes)
```

```python
#Question 3 - Agriculture – Rainfall Adequacy Checker
```

```python
# Function to check rainfall adequency
def rainfall_checker(rainfall_data, required_level):
    #Calculating the total rainfall
    total = 0
    for rain in rainfall_data:
        total = total + rain

    #Calculating average 
    average = total / len(rainfall_data)

    #Checking the condition 
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    
    return average, status


rainfall = [70, 75, 80, 65]
average, status = rainfall_checker(rainfall, 70)

#printing the result
print("Average Rainfall:", average)
print("Rainfall Status:", status)
```

```python
#Question 4 - Social Media – Duplicate Account Detection
```

```python
# Function to check duplicate accounts 
def detect_duplicates(usernames):
    # Converting this list to a set 
    unique_users = set(usernames)

    # Checking for duplicates
    if len(unique_users) < len(usernames):
        return "Yes"
    else:
        return "No"


users_IDs = ["user1", "user2", "user1", "user3"]
result = detect_duplicates(users_IDs)

print("Duplicate Accounts Found:", result)
```

```python
#Question 6 - Agriculture – Premium Crop Price Filter
```

```python
# Function to filter out premium crops
def filter_crop_price(prices):
    # To store all the premium prices
    premium_crop = [] 

    # Checking for the condition of price for above 2000
    for price in prices:
        if price > 2000:
            premium_crop.append(price)
    #printint the list of premium crops
    print(f"Premium Crops: {premium_crop}")

prices = [1500, 2500, 1800, 3200]
filter_crop_price(prices)
```

```python
#Question 7 - System Monitoring – Application Health Checker
```

```python
# Function for system monitoring
def check_app_health(errors):
    
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    
    # Printing the results
    print("Error Count:", errors)
    print("System Status:", status)


# user input for no of errors
errors = int(input("Enter number of errors: "))

# function calling
check_app_health(errors)
```

```python
#Question 9 - E-Learning – Student Attendance Eligibility System
```

```python
# Function to check students attendance eligibility
def attendance_eligibility(attendance_list):
    present = 0
    for attendance in attendance_list:
        if attendance == "P":
            present += 1

    # Calculating the percentage
    percentage = (present / len(attendance_list)) * 100

    # Checking the condition 
    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    # Printing the result
    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)

# Functing call
attendance_eligibility(["P", "P", "A", "P", "P"])
```
