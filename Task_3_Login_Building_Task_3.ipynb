```python
#Question 1 -  Employee Performance Bonus Eligibility
```

```python
# Given input in dictionary form
employees = {
    "Ravi" : 92,
    "Anita" : 88,
    "Kiran" : 92,
    "Suresh" : 85
}

# Finding highst score 
highest_score = max(employees.values())

# Get name of employees with the highest score
top_performers = []
for name, score in employees.items():
    if score == highest_score:      
        top_performers.append(name)

# Printing the result 
print("Top Performers Eligible for Bonus: ", " , ".join(top_performers), "(Score: ", highest_score, ")")
```

```python
#Question 2 - Search Query Keyword Analysis
```

```python
#Input string given
query = "Buy mobile phone buy phone online"

# Converting the string to lowercase
query = query.lower()

# Removing all punctuation marks
punctuations = ".,!?:;"
for p in punctuations:
    query = query.replace(p, "")

# Splitting the query into words 
words = query.split()

# Counting frequency of each words
word_count = {}
for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

# finding the word occuring more than one time
keywords = {}
for word, count in word_count.items():
    if count > 1:
        keywords[word] = count

# Printing the result 
print(keywords)
```

```python
#Question 3 - Sensor Data Validation
```

```python
# Given input readings
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# empty list to store readings
valid_readings = []

for i, value in enumerate(sensor_readings):
    if value % 2 == 0:  # Check if even
        valid_readings.append((i, value))

# PRinting the final results
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
#Question 4 - Email Domain Usage Analysis
```

```python
# Given emails input
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# For storing domain counts 
domain_count = {}

# counting 
for email in emails:
    domain = email.split("@")[1]
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

# Total no of mails
total_mails = len(emails)

# Percentage calculating
for domain, count in domain_count.items():
    percentage = (count / total_mails) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
#Question 5 - Sales Spike Detection
```

```python
# given sales list 
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculating the average
average_sales = sum(sales) / len(sales)

# More than 30% above average
threshold = average_sales * 1.3

for day, sale in enumerate(sales, start=1):
    if sale > threshold:
        print(f"Day {day}: {sale}")
```

```python
#Question 6 - Duplicate User ID Detection
```

```python
# Given user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# to count number of times occuring
count_id = {}

# Count each user id
for user in user_ids:
    if user in count_id:
        count_id[user] += 1
    else:
        count_id[user] = 1

# Printig results of ids occuring more than ones
for user, count in count_id.items():
    if count > 1:
        print(user, "->", count, "times")
```
