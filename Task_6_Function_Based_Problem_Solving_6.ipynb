```python
#Function to manage parking 
def manage_parking(capacity, logs):
    current_count = 0
    peak_usage = 0

    for log in logs:
        if log == "IN":
            current_count += 1
        elif log == "OUT" and current_count > 0:
            current_count -= 1

        # Checking the peak usage
        if current_count > peak_usage:
            peak_usage = current_count

    print("Currently Parked Vehicles:", current_count)

    # Checking the parking limit
    if current_count > capacity:
        print("Parking Status: Exceeded Capacity ")
    else:
        print("Parking Status: Available")

    print("Peak Parking Usage:", peak_usage)


# Inputs given 
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

#Function call
manage_parking(capacity, logs)
```

```python
#Question 2 - Online Food Delivery Time Estimator
```

```python
# Function to estimate delivery time 
def Estimate_delivery_time(distance, traffic_level, weather_condition):
    # (Distance / 30km per hour) * 60 minutes 
    base_time = (distance / 30) * 60

    # Checking the traffic delay
    if traffic_level.lower() == "high":
        base_time += 25
    elif traffic_level.lower() == "medium":
        base_time += 10

    # Checking the weather delay
    if weather_condition.lower() == "rainy":
        base_time += 14
    elif weather_condition.lower() == "stormy":
        base_time += 25

    # Final time estimated 
    final_time = round(base_time)

    print("Estimated Delivery Time:", final_time, "minutes")


# Function Call
Estimate_delivery_time(8, "High", "Rainy")
```

```python
#Question 3 - Movie Theatre Seat Occupancy Analyzer
```

```python
# Function to analyse seat occupancy
def analyze_seat_occupancy(total_seats, booked_seats):

    # Calculating seat occupancy percentage
    occupancy = (len(booked_seats) / total_seats) * 100

    # Checking for theatre status 
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"

    # Suggesting additional shows when the demand is high
    if occupancy >= 90:
        suggestion = "Consider opening additional shows."
    else:
        suggestion = "No additional shows required."

    # Printing the result
    print("Occupancy:", round(occupancy), "%")
    print("Show Status:", status)
    print("Suggestion:", suggestion)


# Inputs given 
total_seats = 200
booked_seats = [1] * 150

# Function Call
analyze_seat_occupancy(total_seats, booked_seats)
```

```python
#Question 4 - Cloud Server Load Classification System
```

```python
# Function to classify the server load 
def classify_server_load(cpu_readings):

    # Calculating the average 
    average = sum(cpu_readings) / len(cpu_readings)

    # classifying the server status based on condition
    if average < 50:
        status = "Normal"
    elif 50 <= average <= 80:
        status = "Warning"
    else:
        status = "Critical"

    # Printing the result
    print("Average CPU Load:", round(average),"%")
    print("Server Status:", status)

# Inputs given 
cpu_readings = [45, 60, 70, 85, 90]

# Function Call
classify_server_load(cpu_readings)
```

```python
#Question 5 - Smart Classroom Resource Usage Monitor
```

```python
# Function to monitor resoure usage
def monitor_resource_usage(resource_data):
    overused = []
    
    # Setting the limit of overuse 
    limit = 8

    # Checking the limit if its greater the the given limit
    for resource, hours in resource_data.items():
        if hours > limit:
            overused.append(resource)

    # Printing the result 
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Inputs given 
usage_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

# Function calling 
monitor_resource_usage(usage_data)
```

```python
# Function to control online event registration capacity 
def control_event_registration(event_capacity, total_registrations):

    confirmed = 0
    waitlist = 0

    # Seat confirming based on availability condition
    for number in range(total_registrations):
        if confirmed < event_capacity:
            confirmed += 1
        else:
            waitlist += 1

    #  updating the registration status based on the confirmed registrations  
    if confirmed >= event_capacity:
        registration_status = "Closed"
    else:
        registration_status = "Open"

    # Printing the result
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", registration_status)


# Given input
capacity = 100
registrations = 105

# Fnction calling
control_event_registration(capacity, registrations)
```
