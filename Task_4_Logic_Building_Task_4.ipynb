```python
# Function to check if its less than 50 and valid plan 
def validate_recharge(amount):
    # List of plans given 
    valid_plans = [199, 299, 399, 599]
    
    # Checking for minimum amount 
    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False
    
    # Checking for valid plans valid plans
    elif amount not in valid_plans:
        print("Invalid plan selected.")
        print("Available plans are:", valid_plans)
        return False
    
    else:
        print("Recharge successful with plan ₹", amount)
        return True

# main function 
while True:
    amount = int(input("Enter recharge amount: "))
    if validate_recharge(amount):
        break
    else:
        print("Please try again\n")
```

```python
#Question 2 -  Inventory Reorder Alert System
```

```python
#function to check inventory status
def check_inventory(products):
    print("Inventory Status")
    print( "=" * 22)
    for product, stock in products.items():
        if stock < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")

#main function
if __name__ == "__main__":
  products = {
        "Wheat": 20,
        "Sugar": 10,
        "Millets": 8,
        "Rice": 25
    }
  check_inventory(products)
```

```python
#Question 3 - Student Result Processing System
```

```python
# function to calculate avg marks 
def process_result(marks):
    total = 0
    
    # calculating total marks
    for mark in marks:
        total += mark
    
    # average
    average = total / len(marks)
    
    print("Average Marks:", average)
    
    # Checking condition
    if average >= 50:
        return "Pass"
    else:
        return "Fail"


# Main function
if __name__ == "__main__":
    
    # list of students marks given 
    student_marks = [60, 45, 55, 70, 40]
    
    # function call
    result = process_result(student_marks)
    
    # Printing the result 
    print("Result:", result)
```

```python
#Question 4 - Cab Fare Estimator with Retry Option
```

```python
# Function to calculate fare
def calculate_fare(distance, peak):
    base_fare = 50
    per_km = 12
    fare = base_fare + (distance * per_km)
    # Add peak charge
    if peak == "yes":
        fare = fare * 1.25

    return fare
    
# Main function
while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Peak hour? (yes/no): ").lower()

    if distance <= 0:
        print("Invalid distance. Try again\n")
        continue

    total_fare = calculate_fare(distance, peak)
    print("Total Fare: ₹", int(total_fare))
    break
```

```python
#Question 5 - Employee Attendance Eligibility Checker
```

```python
# Function for checking eligibility
def check_eligibility(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present += 1

    # Calculating attendance percentage
    percentage = (present / len(attendance)) * 100
    print("Attendance Percentage:", percentage)
    
    # Checking the condition minimum 75%
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

attendance_list = ["P", "P", "P", "A", "P"]
print("Status:", check_eligibility(attendance_list))
```

```python
#Question 6 - Password Strength Checker
```

```python
# Function for checking passwords
def check_password(password):
  has_digit = False
  has_special = False
  special_chars = "@#$"
    
  # checking for digit and special characters 
  for ch in password:
      if ch.isdigit():
          has_digit = True
      if ch in special_chars:
          has_special = True
          
  # checking the lenght and digit and special characters 
  if len(password) >= 8 and has_digit and has_special:
        return "Strong Password"
  else:
        return "Weak Password"

#main function
check= input("Enter password: ")
result = check_password(check)
print(result)
```
